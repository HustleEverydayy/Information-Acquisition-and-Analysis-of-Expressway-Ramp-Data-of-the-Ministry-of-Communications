```python
import pandas as pd

SrcDir = "../SRC/"
SrcFile = 'part-r-00000_TDCS_201611_201910_03F1779S_03F2129S_G8_Key_TotalTime_TimeIntervals_TFCFLength_ClassFrequency_1_0'
SrcLocation = SrcDir +SrcFile

#pd.read_csv(SrcLocation)
#pd.read_csv(SrcLocation,sep='\#\#')
#pd.read_csv(SrcLocation,sep='\#\#|\t')
TDCS_MRP_data = pd.read_csv(SrcLocation,sep='\#\#|\t', header=None)
#TDCS_MRP_data
TDCS_MRP_data.columns = ["Hour","TotalTravelTime","TravelTimeIntervals","TF","DF","NumOfGantries","ClassFrequencyDistribution"]
TDCS_MRP_data
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html
#https://www.tutorialspoint.com/python_pandas/python_pandas_groupby.htm
#https://zhuanlan.zhihu.com/p/101284491

TDCS_MRP_data.groupby(['TravelTimeIntervals']).agg({'TotalTravelTime':'median', 'TF':'sum'})
```

```python
TDCS_MRP_data_GroupTTT_SumTF = TDCS_MRP_data.groupby(['TotalTravelTime']).agg({ 'TF':'sum'})
TDCS_MRP_data_GroupTTT_SumTF
```

```python
TDCS_MRP_data_GroupTTT_SumTF.index.tolist()
```

```python
TDCS_MRP_data_GroupTTT_SumTF['TF'].tolist()
```

```python
TotalTravelTime_X_TF  = TDCS_MRP_data_GroupTTT_SumTF.index * TDCS_MRP_data_GroupTTT_SumTF['TF']
TotalTravelTime_X_TF
```

```python
Total_Trip_TravelTime = TotalTravelTime_X_TF.sum()
Total_Trip_TravelTime
```

```python
Total_TF = TDCS_MRP_data_GroupTTT_SumTF.sum()
Total_TF
```

```python
Total_Trip_TravelTime/Total_TF
```

```python
# How to determine the range of normal and abnormal according to Normal Distribution ?
# https://www.investopedia.com/terms/n/normaldistribution.asp

import matplotlib.pyplot as plt

x = TDCS_MRP_data_GroupTTT_SumTF.index.tolist()
y = TDCS_MRP_data_GroupTTT_SumTF['TF'].tolist()
plt.bar(x,y)
plt.xlabel('TotalTravelTime')
plt.ylabel('# of Trips')
plt.title('part-r-00000_TDCS_201611_201910_03F1779S_03F2129S_G8')
```
