```python
import pandas as pd

SrcDir = "../SRC/"
SrcFile = 'part-r-00000_TDCS_201611_201910_03F1779S_03F2129S_G8_Key_TotalTime_TimeIntervals_TFCFLength_ClassFrequency_1_0'
SrcLocation = SrcDir +SrcFile

#pd.read_csv(SrcLocation)
#pd.read_csv(SrcLocation,sep='\#\#')
#pd.read_csv(SrcLocation,sep='\#\#|\t')
data = pd.read_csv(SrcLocation,sep='\#\#|\t', header=None)
```

```python
data
```

```python
data.columns = ["Hour","TotalTravelTime","TravelTimeIntervals","TF","DF","NumOfGantries","ClassFrequencyDistribution"]
```

```python
data
```

```python
data.describe()
```

```python

TF_TotalTravelTime_Distribution = data[['TotalTravelTime','TF']].groupby(['TotalTravelTime']).sum()
TF_TotalTravelTime_Distribution
```

```python
TF_TotalTravelTime_Distribution.describe()
```

```python
TF_TotalTravelTime_Distribution.plot()
```

```python
import matplotlib.pyplot as plt

x = TF_TotalTravelTime_Distribution.index
y = TF_TotalTravelTime_Distribution['TF']
#x

#y
fig1 = plt.figure(1,figsize=(8,6), dpi=80)
plt.plot(x,y,'r-',lw=2,label='TF')
plt.xlabel('TotalTravelTime')
plt.ylabel('Frequency')

plt.title('The frequency distributions of Vehiles traveling from 03F1779S to 03F2129S (8 Gangtries) according to TotalTravelTime ')
plt.grid(axis='both')
plt.legend(loc='best')
plt.axhline(y=0, color='black')

plt.savefig('03F1779S_03F2129S_8_Gantries.png')
plt.show()
          
          
          
```
